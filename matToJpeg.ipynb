```python
cnt1 = 0
cnt2 = 0
cnt3 = 0

for i in range(3064):
    if i == 0: 
        continue
    
    f = h5py.File('../brainTumorDataPublic/'+str(i)+'.mat','r')
    data = f.get('cjdata/image')
    data = np.array(data)
    
    label = f.get('cjdata/label')
    
    if(label.value == 1):
        scipy.misc.imsave('../Data/1/'+str(cnt1)+'.jpg', data)
        cnt1 = cnt1 + 1
    
    elif(label.value == 2):
        scipy.misc.imsave('../Data/2/'+str(cnt2)+'.jpg', data)
        cnt2 = cnt2 + 1
        
    elif(label.value == 3):
        scipy.misc.imsave('../Data/3/'+str(cnt3)+'.jpg', data)
        cnt3 = cnt3 + 1
```
